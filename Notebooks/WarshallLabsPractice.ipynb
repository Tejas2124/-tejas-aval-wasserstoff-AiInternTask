```python
import fitz
from fitz import open as fitz_open



def is_scanned_pdf(pdf_path):
    """
    Checks if a PDF is likely scanned based on its content.

    Args:
        pdf_path (str): Path to the PDF file.

    Returns:
        bool: True if likely scanned, False otherwise.
    """
    try:
        doc = fitz_open(pdf_path)
        total_text_area = 0
        total_page_area = 0

        for page in doc:
            total_page_area += abs(page.rect)
            text_blocks = page.get_text("blocks")
            text_area = sum(abs(fitz.Rect(b[:4])) for b in text_blocks)
            total_text_area += text_area

        doc.close()

        text_percentage = total_text_area / total_page_area if total_page_area > 0 else 0
        return text_percentage < 0.01  #Threshold can be adjusted

    except Exception as e:
       print(f"An error occurred: {e}")
       return False
```

## Paragraph wise splitting with reg exp

```python
import re

def clean_paragraphs(raw_text):
    lines = raw_text.splitlines()
    paragraphs = []
    paragraph = ''

    for line in lines:
        line = line.strip()
        if not line:
            if paragraph:
                paragraphs.append(paragraph.strip())
                paragraph = ''
        else:
            if paragraph and not paragraph.endswith(('.', '!', '?')):
                paragraph += ' ' + line
            else:
                paragraph += (' ' if paragraph else '') + line

    if paragraph:
        paragraphs.append(paragraph.strip())

    return paragraphs

# Usage
paragraphs = clean_paragraphs(result[0])
for i, para in enumerate(paragraphs, 1):
    print(f"Paragraph {i}:\n{para}\n")
```
